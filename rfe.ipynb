```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
```

```python
tipo = "log"
data_5d  = pd.read_csv('./in/copper_returns_5d.csv')
data_10d = pd.read_csv('./in/copper_returns_10d.csv')
data_20d = pd.read_csv('./in/copper_returns_20d.csv')
data_5d_log  = pd.read_csv('./in/copper_log_returns_5d.csv')
data_10d_log = pd.read_csv('./in/copper_log_returns_10d.csv')
data_20d_log = pd.read_csv('./in/copper_log_returns_20d.csv')
data_5d.head()
```

```python
features_5d  = data_5d .drop(['Date', 'LMCADY_acu_5d',  'LMCADY_std_5d'] , axis=1)
features_10d = data_10d.drop(['Date', 'LMCADY_acu_10d', 'LMCADY_std_10d'], axis=1)
features_20d = data_20d.drop(['Date', 'LMCADY_acu_20d', 'LMCADY_std_20d'], axis=1)

target_5d  = data_5d ['LMCADY_acu_5d' ]
target_10d = data_10d['LMCADY_acu_10d']
target_20d = data_20d['LMCADY_acu_20d']

X_train_5d, X_test_5d, y_train_5d, y_test_5d = train_test_split(features_5d, target_5d, test_size=0.1, shuffle=False, random_state=42)
X_train_10d, X_test_10d, y_train_10d, y_test_10d = train_test_split(features_10d, target_10d, test_size=0.1, shuffle=False, random_state=42)
X_train_20d, X_test_20d, y_train_20d, y_test_20d = train_test_split(features_20d, target_20d, test_size=0.1, shuffle=False, random_state=42)
```

```python
features_5d_log  = data_5d_log .drop(['Date', 'LMCADY_acu_5d_log',  'LMCADY_std_5d_log'] , axis=1)
features_10d_log = data_10d_log.drop(['Date', 'LMCADY_acu_10d_log', 'LMCADY_std_10d_log'], axis=1)
features_20d_log = data_20d_log.drop(['Date', 'LMCADY_acu_20d_log', 'LMCADY_std_20d_log'], axis=1)

target_5d_log  = data_5d_log ['LMCADY_acu_5d_log' ]
target_10d_log = data_10d_log['LMCADY_acu_10d_log']
target_20d_log = data_20d_log['LMCADY_acu_20d_log']

X_train_5d_log, X_test_5d_log, y_train_5d_log, y_test_5d_log = train_test_split(features_5d_log, target_5d_log, test_size=0.1, shuffle=False, random_state=42)
X_train_10d_log, X_test_10d_log, y_train_10d_log, y_test_10d_log = train_test_split(features_10d_log, target_10d_log, test_size=0.1, shuffle=False, random_state=42)
X_train_20d_log, X_test_20d_log, y_train_20d_log, y_test_20d_log = train_test_split(features_20d_log, target_20d_log, test_size=0.1, shuffle=False, random_state=42)
```

```python
rf_5d  = RandomForestRegressor(n_estimators=100, random_state=5)
rf_10d = RandomForestRegressor(n_estimators=100, random_state=10)
rf_20d = RandomForestRegressor(n_estimators=100, random_state=20)
```

```python
rf_5d_log  = RandomForestRegressor(n_estimators=100, random_state=5)
rf_10d_log = RandomForestRegressor(n_estimators=100, random_state=10)
rf_20d_log = RandomForestRegressor(n_estimators=100, random_state=20)
```

```python
n_features = 5
rfe_5d = RFE(estimator=rf_5d, n_features_to_select=n_features, step=1)
rfe_5d.fit(X_train_5d, y_train_5d)
```

```python
rfe_10d = RFE(estimator=rf_10d, n_features_to_select=n_features, step=1)
rfe_10d.fit(X_train_10d, y_train_10d)
```

```python
rfe_20d = RFE(estimator=rf_20d, n_features_to_select=n_features, step=1)
rfe_20d.fit(X_train_20d, y_train_20d)
```

```python
rfe_5d_log = RFE(estimator=rf_5d_log, n_features_to_select=n_features, step=1)
rfe_5d_log.fit(X_train_5d_log, y_train_5d_log)
```

```python
rfe_10d_log = RFE(estimator=rf_10d_log, n_features_to_select=n_features, step=1)
rfe_10d_log.fit(X_train_10d_log, y_train_10d_log)
```

```python
rfe_20d_log = RFE(estimator=rf_20d_log, n_features_to_select=n_features, step=1)
rfe_20d_log.fit(X_train_20d_log, y_train_20d_log)
```

```python
selected_features_5d = pd.DataFrame({'Variable 5d':features_5d.columns, 'Importancia':rfe_5d.ranking_})
selected_features_5d.sort_values(by="Importancia", ascending=True)
```

```python
selected_features_10d = pd.DataFrame({'Variable 10d':features_10d.columns, 'Importancia':rfe_10d.ranking_})
selected_features_10d.sort_values(by="Importancia", ascending=True)
```
